```python
#Importamos bibliotecas y realizamos conexion a la base de datos
from pymongo import MongoClient
import pymongo
import json
import matplotlib.pyplot as plt
from bson.json_util import dumps, loads

# Conéctate al servidor MongoDB
client = MongoClient('mongodb://localhost:27017/')

# Selecciona la base de datos
db = client['ejercicioclase']

# Selecciona la colección
collection = db['restaurante']

# Realiza una consulta para verificar la conexión
documento = collection.find_one()
print(documento)
```

```python
#Consulta 22: Restaurantes con la calificacion A y puntuaje 11 en la fecha especifica ISODate "2014-08-11T00:00:00Z" 
from datetime import datetime
specific_date = datetime(2014, 8, 11)
consulta_22 = collection.find(
    {"grades": {"$elemMatch": {"grade": "A", "score": 11, "date": specific_date}}},
    {"restaurant_id": 1, "name": 1, "grades": 1})
for doc in consulta_22:
    print(doc)
```

```python
#Consulta 30: Restaurantes con puntuacion divisible por 7
cursor = collection.find(
    {"grades": {"$elemMatch": {"score": {"$mod": [7,0]}}}},
    {"restaurant_id": 1, "name": 1, "grades": 1})
documents = list(cursor)
for doc in documents:
    print(doc)
```

```python
#Consulta 31: Restaurantes con 'mon' en el nombre
cursor = collection.find(
    {"name": {"$regex": ".*mon.*", "$options": "i"}},
    {"name": 1, "borough": 1, "address.coord": 1, "cuisine": 1})

documents = list(cursor)
for doc in documents:
    print(doc)
```

```python
#Consulta 32: Restaurantes con nombre que empieza con 'Mad'
cursor = collection.find(
    {"name": {"$regex": "^Mad"}},
    {"name": 1, "borough": 1, "address.coord": 1, "cuisine": 1})
documents = list(cursor)
for doc in documents:
    print(doc)
```

```python
#Visualizacion de datos
"""
usamos matplotlib para crear graficos que muestren la distribucion de datos.
"""
#Distribucion de 'Borough'
#Grafico de barras que muestra la cantidad de restaurantes por 'borough'.
boroughs = collection.aggregate([
    {"$group": {"_id": "$borough", "count": {"$sum": 1}}}
])
boroughs = list(boroughs)
labels = [b["_id"] for b in boroughs]
sizes = [b["count"] for b in boroughs]
plt.figure(figsize=(10,5))
plt.bar(labels, sizes)
plt.title("Distribucion de Boroughs")
plt.xlabel("Boroughs")
plt.ylabel("Cantidad")
plt.show()
```

```python
#Añadir un nuevo campo 'avg_score' a cada documento

for doc in collection.find():
    grades = doc.get("grades", [])
    if grades:
        total_score = sum(grade["score"] for grade in grades if "score" in grade)
        avg_score = total_score / len(grades)
                                
    else:
        avg_score = None
    
    #Actualizamos el documento con el nuevo campo 'avg_score'
    collection.update_one({"_id": doc["_id"]}, {"$set": {"avg_score": avg_score}})

print("Actualizacion completada: Campo 'avg_score' añadido")                                
```
